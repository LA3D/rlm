# Knowledge Fabric: Definition

A **knowledge fabric** is an interconnected network of SPARQL endpoints, each exposing structured semantic data through shared ontologies and cross-references.

## Key Characteristics

1. **Distributed**: Data lives at specialized endpoints (proteins, compounds, pathways, etc.)
2. **Linked**: Entities reference each other via URIs and `rdfs:seeAlso`
3. **Self-describing**: Ontologies define structure; examples show usage patterns
4. **Traversable**: Agents follow links across endpoints to answer complex questions

## How It Enables Agentic Query

An agent like me can:
- **Discover** structure by reading ontologies
- **Navigate** by following cross-references between endpoints  
- **Federate** queries to combine data from multiple sources
- **Reason** using ontology hierarchies (subclasses, property paths)

## Core Methodology: Follow Your Nose

### Step 1: Understand the Ontology
- Read ontology files (`.ttl`, `.owl`) to discover classes and properties
- Check for agent guides or documentation (e.g., `AGENT_GUIDE.md`)
- Review example queries to understand patterns

Ontologies are in the ./ontology directory in the local file system. Uniprot and wikipathways have both the core ontology as well as examples.
Specifically:
- ./ontology/uniprot 
- ./ontology/wikipathways
- ./sio
- ./pubchem

### Step 2: Construct Queries Using Ontology Knowledge
- Use discovered classes and properties
- Leverage hierarchies (`rdfs:subClassOf`, property paths)
- Apply patterns from examples

### Step 3: Reflect on Results
- Examine what properties/relationships exist
- Follow links (`rdfs:seeAlso`, cross-references)
- Iterate and refine queries

## Key Principles

1. **Start broad, then filter**: Explore what exists before constraining
2. **Use ontology structure**: Hierarchies let you query at any level
3. **Check examples first**: They reveal endpoint-specific patterns
4. **Follow cross-references**: Connect across endpoints
5. **Federate when needed**: Combine data from multiple sources

**Gene:** DNM1L

**Main Function:** 
DNM1L is crucial for **mitochondrial and peroxisomal division**. It wraps around organelles and uses GTP energy to pinch them apart, creating new mitochondria through a process called fission.

**Key Roles:**
- Essential for normal brain development, especially the cerebellum
- Required for mitochondrial division during cell division (mitosis)
- Helps neurons survive by preventing oxidative damage
- Involved in programmed cell death (apoptosis and necrosis)
- Regulates circadian rhythm through mitochondrial ATP production

**Where it's found:**
- Mitochondrial outer membrane (primary location)
- Peroxisomes
- Synaptic vesicles
- Golgi apparatus
- Cytosol

**Connection to Alzheimer's:** In AD, amyloid-beta causes excessive modification of DNM1L, leading to too much mitochondrial fission, which damages synapses and neurons.
