# Knowledge Fabric: Definition

A **knowledge fabric** is an interconnected network of SPARQL endpoints, each exposing structured semantic data through shared ontologies and cross-references.

## Key Characteristics

1. **Distributed**: Data lives at specialized endpoints (proteins, compounds, pathways, etc.)
2. **Linked**: Entities reference each other via URIs and `rdfs:seeAlso`
3. **Self-describing**: Ontologies define structure; examples show usage patterns
4. **Traversable**: Agents follow links across endpoints to answer complex questions

## How It Enables Agentic Query

An agent like me can:
- **Discover** structure by reading ontologies
- **Navigate** by following cross-references between endpoints  
- **Federate** queries to combine data from multiple sources
- **Reason** using ontology hierarchies (subclasses, property paths)

## Core Methodology: Follow Your Nose

### Step 1: Understand the Ontology
- Read ontology files (`.ttl`, `.owl`) to discover classes and properties
- Check for agent guides or documentation (e.g., `AGENT_GUIDE.md`)
- Review example queries to understand patterns

Ontologies are in the ./ontology directory in the local file system. Uniprot and wikipathways have both the core ontology as well as examples.
Specifically:
- ./ontology/uniprot 
- ./ontology/wikipathways
- ./sio
- ./pubchem

### Step 2: Construct Queries Using Ontology Knowledge
- Use discovered classes and properties
- Leverage hierarchies (`rdfs:subClassOf`, property paths)
- Apply patterns from examples

### Step 3: Reflect on Results
- Examine what properties/relationships exist
- Follow links (`rdfs:seeAlso`, cross-references)
- Iterate and refine queries

## Key Principles

1. **Start broad, then filter**: Explore what exists before constraining
2. **Use ontology structure**: Hierarchies let you query at any level
3. **Check examples first**: They reveal endpoint-specific patterns
4. **Follow cross-references**: Connect across endpoints
5. **Federate when needed**: Combine data from multiple sources

## Core Function & Disease Association

**Primary Role**: DNM1L is a GTPase that mediates **mitochondrial and peroxisomal fission** - essentially acting as molecular "scissors" that divide these organelles. It wraps around organelle membranes and constricts them through GTP hydrolysis.

**Alzheimer's Connection**: The protein is associated with Alzheimer's disease through a specific mechanism: amyloid-beta induces increased S-nitrosylation of DNM1L, which triggers excessive mitochondrial fission, leading to synaptic loss and neuronal damage. This is a key pathological mechanism in AD.

## Cellular Functions

DNM1L has multiple critical roles:
- **Mitochondrial division** during mitosis and normal cellular processes
- **Peroxisomal division**
- **Endocytic vesicle formation**
- **Synaptic vesicle membrane dynamics**
- **Programmed necrosis execution**
- **Circadian control** of mitochondrial ATP production
- **Brain development**, particularly cerebellar development
- **Neuronal survival** by suppressing oxidative damage in postmitotic neurons

## Subcellular Location

DNM1L is highly dynamic in its localization:
- Mainly **cytosolic** but recruited to membranes
- **Mitochondrial outer membrane** at division sites (recruited by receptors like MFF, MIEF1, MIEF2)
- **Peroxisomal membranes** (recruited by PEX11B)
- **Endoplasmic reticulum** tubules
- **Golgi complex** (in some cell types)
- **Clathrin-coated pits**
- **Synaptic vesicles**

## Expression Pattern

- **Ubiquitously expressed** with highest levels in skeletal muscle, heart, kidney, and **brain**
- Has **6 isoforms** with tissue-specific expression:
  - Isoform 1: brain-specific
  - Isoform 6: expressed in neurons
  - Others vary in testis, skeletal muscle, liver, heart, kidney

## Protein Structure

- Contains a **GED (GTPase effector domain)** that:
  - Folds back to interact with the GTP-binding domain
  - Is critical for activating GTPase activity
  - Enables DNM1L dimerization
- Has GTP-binding and middle domains
- Forms oligomeric structures that assemble into membrane-wrapping rings

## Key Structural Features for Function

1. **GTPase domain** - Hydrolyzes GTP to drive membrane constriction
2. **Middle domain** - Enables oligomerization for ring formation
3. **GED domain** - Folds back to activate GTPase; enables dimerization
4. **Disordered B region** - Contains regulatory phosphorylation sites

The protein assembles into **helical rings** around mitochondrial membranes, and GTP hydrolysis drives membrane constriction and fission.
